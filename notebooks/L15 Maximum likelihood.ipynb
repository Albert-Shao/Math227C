# Maximum likelihood estimation
Given a single data point, $T$, one strategy for inferring the parameter $\lambda$ is to choose the parameter that maximizes the likelihood.

For an exponential random variable,
$$L(\lambda) = \lambda e^{-\lambda T}$$
is maximized when
$$\frac{dL}{d\lambda}=0 \implies \lambda=1/T.$$

You could also find the max numerically.

For $N$ i.i.d. random variables, the likelihood is the product of likelihoods (just like the probability density). So, e.g., for $N$ Gaussian random variables $X_i$,
$$L(\mu,\sigma) = \Pi_{i=1}^N p(X_i;\sigma,\mu)$$
